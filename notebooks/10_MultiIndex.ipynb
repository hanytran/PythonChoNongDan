```python
import pandas as pd
import numpy as np

index = pd.MultiIndex.from_product([[2019, 2020], [1, 2, 3, 4]],
                                   names=['year', 'quarter'])

columns = pd.MultiIndex.from_product([['Anna', 'Bill', 'Carl'], ['P1', 'P2']],
                                     names=['name', 'type'])

# mock some data
data = np.random.randint(50, 200, (8, 6))

# create the DataFrame
df = pd.DataFrame(data, index=index, columns=columns)
df
```
